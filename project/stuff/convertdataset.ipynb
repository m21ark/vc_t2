```python
# load a pandas csv

import pandas as pd
import ast

df = pd.read_csv("data/values.csv")

df.head()
```

```python
# for each row, create a txt file with the name "id".txt and write the bb_list value to the file

for index, row in df.iterrows():
    with open(f"parsed/{row['id']}.txt", "w") as f:
        
        bb = ast.literal_eval(row['bb_list'])
        
        for b in bb:
            print(b)
            x_center = (b[0] + b[2]) / 2
            y_center = (b[1] + b[3]) / 2
            width = b[2] - b[0]
            height = b[3] - b[1]
            
            f.write(f"0 {x_center} {y_center} {width} {height}\n")
    

```
